# POS Tagging
Parts of speech tags are the properties of the words, which define their main context, functions, and usage in a sentence.
image.png

```python
import nltk
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
```

```python
nltk.help.upenn_tagset()
```

```python
sentence = word_tokenize("whatever the world is a great place")
nltk.pos_tag(sentence)
```

```python
sentence = word_tokenize("applicant is removed from applicant list of the job ")
nltk.pos_tag(sentence)
```
